```python
# def wc():

seeds = [42, 26092020, 24032022]    
path_test = str(os.getcwd()) + "/result/seed" + str(42) + "/" + "test" + "/"
dir_list_test = [f for f in os.listdir(path_test)]

word_candidates_emot = []
word_candidates_sent = []

for seed in seeds:
    for f in dir_list_test:
        df = pd.read_csv(os.getcwd() + r'/result/seed'+str(seed)+"/test/"+str(f))

        if "emotion" in f:
            for el in df['translated_word(s)'].dropna().values:
                el = ast.literal_eval(el)
                word_candidates_emot.append([(k, v) for k, v in el.items()])
        if "sentiment" in f:
            for el in df['translated_word(s)'].dropna().values:
                el = ast.literal_eval(el)
                word_candidates_sent.append([(k, v) for k, v in el.items()])

merged_emot = list(itertools.chain(*word_candidates_emot))
merged_sent = list(itertools.chain(*word_candidates_sent))

print(len(merged_emot))
print(len(merged_sent))
```
